**Python for Loop with Python range()**  
A range is a series of values between two intervals.



```python
a = range(4)

for i in a:
    print(i)
```

**Python for loop with the break Statement**  
break statement will stop the loop before it has looped through all the items:

```python
a = ["the","banana","apple","python"]
for x in a:
  print(x)
  if x == "banana":
    break
```

```python
a = ["the","banana","apple","python"]
for x in a:
  
  if x == "banana":
    break
  print(x)
```

**Python for loop with continue statement**  
Continue statement will stop the current iteration of the loop, and continue.

```python
a = ["the","banana","apple","python"]
for x in a:
  print(x)
  if x == "banana":
    continue
```

```python
a = ["the","banana","apple","python"]
for x in a:
  
  if x == "banana":
    continue
  print(x)
```

**Python for loop with Using else Statement**  
else statement associated with a loop statement

If the else statement is used with a for loop, the else statement is executed when the loop has exhausted iterating the list.

```python
for a in range(9):
  print(a)
else:
  print("completed")
```

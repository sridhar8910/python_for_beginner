#**Python**
In this session we learn about Tuples and
**Tuple**

A tuple is a collection of objects which ordered and immutable. Tuples are sequences, just like lists. The differences between tuples and lists are, the tuples cannot be changed unlike lists and tuples use parentheses, whereas lists use square brackets.A tuple is a collection which is ordered and unchangeable.  
**Creating a Tuple**  
A tuple is created by placing all the items (elements) inside parentheses (), separated by commas. 

```python
a = ('physics', 'chemistry', 1997, 2000);
b = (4, 5, 6, 9, 5 );
c = "p", "y", "t", "h", "o", "n";
print(a)
print(b)
print(c)
```

**Create Tuple With One Item**  
To create a tuple with only one item, you have to add a comma after the item, otherwise Python will not recognize it as a tuple.We can use the type() function to know which class a variable or a value belongs to.

```python
a = ("python")
print(type(a))  


b = ("python",)
print(type(b)) 

c = "python",
print(type(c))  
```

**Tuples VS Lists:**  
Similarities:  
 * Functions that can be used for both lists and tuples:len(), max(), min(), sum(), any(), all(), sorted()  
 * Methods that can be used for both lists and tuples:
count(), Index()  
 * Tuples can be stored in lists.  
 * Lists can be stored in tuples.  
 * Both ‘tuples’ and ‘lists’ can be nested.

Differences:
 * Methods that cannot be used for tuples:
append(), insert(), remove(), pop(), clear(), sort(), reverse()
 * we generally use ‘tuples’ for heterogeneous (different) data types and ‘lists’ for homogeneous (similar) data types. 
 * Iterating through a ‘tuple’ is faster than in a ‘list’.  
 * ‘Lists’ are mutable whereas ‘tuples’ are immutable. 
 * Tuples that contain immutable elements can be used as a key for a dictionary.

#**sets**  
> Set is an unordered collection of data types that is iterable, mutable and has no duplicate elements.    

> we create sets by placing all the elements inside curly braces {}, separated by comma.

> A set can have any number of items and they may be of different types (integer, float, tuple, string etc.). But a set cannot have mutable elements like lists, sets or dictionaries as its elements.

```python

#create a list in python 
list1={1,2,3,4,4,5,5,5}
print(list1)
list2={"python","is","eassy ","language"}
print(list2)
list3={1.3,123.32,234.4,24,23.32}
print(list3)
```

**Advantages:**  
* Unique Elements:
 Sets can only contain unique elements, so they can be useful for removing duplicates from a collection of data.  
* Fast Membership Testing: Sets are optimized for fast membership testing, so they can be useful for determining whether a value is in a collection or not.  
* Mathematical Set Operations: Sets support mathematical set operations like union, intersection, and difference, which can be useful for working with sets of data.
* Mutable: Sets are mutable, which means that you can add or remove elements from a set after it has been created.  

**Disadvantages:**  
* Unordered: Sets are unordered, which means that you cannot rely on the order of the data in the set. This can make it difficult to access or process data in a specific order.
* Limited Functionality: Sets have limited functionality compared to lists, as they do not support methods like append() or pop(). This can make it more difficult to modify or manipulate data stored in a set.
* Memory Usage: Sets can consume more memory than lists, especially for small datasets. This is because each element in a set requires additional memory to store a hash value.
* Less Commonly Used: Sets are less commonly used than lists and dictionaries in Python, which means that there may be fewer resources or libraries available for working with them. This can make it more difficult to find solutions to problems or to get help with debugging.

```python
#add
list1={1,2,3,4,4,5,5,5} 
list1.add(34)
print(list1)

list2={"python","is","eassy ","language"}
list2.add(" coding")
print(list2)
```

```python
#clearing the list
list1={1,2,3,4,4,5,5,5} 
list1.clear()
print(list1)
```

```python
#differnce() 
list1={1,2,3,4,4,5,5,5} 
list2={21,323,43,1,2,3,4}

print(list1.difference(list2))
print(list2.difference(list1))
```

```python
#union
list1={1,2,3,4,4,5,5,5} 
list2={21,323,43,1,2,3,4}
print(list1.union(list2))
```

```python
#intersection()
list1={1,2,3,4,4,5,5,5} 
list2={21,323,43,1,2,3,4}
print(list1.intersection(list2))
```
